```python
theta = np.linspace(0, np.pi, 100)
x = np.linspace(0, 10, 1000)
y = np.sin(theta.reshape(-1, 1) * x)
```

```python
plt.figure()
plt.plot(x, y[0])
plt.plot(x, y[1])
plt.plot(x, y[2])
plt.plot(x, y[3])
plt.plot(x, y[4])
plt.show()
```

```python
components = np.linalg.svd(y, full_matrices=False)
```

```python
components[0][:,0] @ components[0][:,1]
```

```python
np.linalg.norm(components[0][:,0])
```

```python
plt.figure()
plt.plot(theta, components[0][:,0])
plt.plot(theta, components[0][:,1])
plt.plot(theta, components[0][:,2])
plt.plot(theta, components[0][:,3])
plt.plot(theta, components[0][:,4])
plt.show()
```

```python
plt.plot(components[1])
plt.yscale('log')
plt.show()
```

```python
plt.plot(x, components[2][0])
plt.plot(x, components[2][1])
plt.plot(x, components[2][2])
plt.plot(x, components[2][3])
plt.plot(x, components[2][4])
plt.show()
```
